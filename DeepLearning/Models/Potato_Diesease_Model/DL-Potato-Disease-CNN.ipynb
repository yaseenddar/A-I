```python
import tensorflow as tf
```

```python
daily_sails_numbers = [12,-432,234,54,100,567,77,88,98]
```

```python
# // built the simple tf dataset from  python list
tf_dataset = tf.data.Dataset.from_tensor_slices(daily_sails_numbers)
```

```python
for data in tf_dataset.take(2):
    print(data.numpy())
```

```python
for data in tf_dataset:
    print(data.numpy())
```

```python
for data in tf_dataset.batch(3):
    print(data.numpy())
```
