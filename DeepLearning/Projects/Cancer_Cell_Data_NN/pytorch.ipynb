```python
class MySimpleModel:
  def __init__(self,X):

    self.weights = torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
    self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)

  def forward(self,X_train):
    mul = torch.matmul(X_train,self.weights) + self.bias
    result = torch.sigmoid(mul);
    return result;

  def loss(self,y_pred,y_true):
    # Ensure y_true has the same shape as y_pred for element-wise operations
    y_true_reshaped = y_true.unsqueeze(1).double() # Add a dimension and cast to double
    epsilon = 1e-7 # Small value to prevent log(0) issues
    loss = - (y_true_reshaped * torch.log(y_pred + epsilon) + (1 - y_true_reshaped) * torch.log(1 - y_pred + epsilon)).mean()
    return loss;

  def train(self,X):
    for epoch in range(epochs):

      y_pred = self.forward(X)
      loss = self.loss(y_pred,X)

      # now we auto gran the loss to calculate the gradients
      loss.backward()

      # now our weights and bias are updated when backward
      with torch.no_grad():
        # Use in-place operations to update weights and bias
        self.weights -= lr*self.weights.grad
        self.bias -= lr*self.bias.grad
        y_pred = (self.forward(X) > 0.5).float()

      #  zeor the grads because it accumulates teh values
      self.weights.grad.zero_()
      self.bias.grad.zero_()

      print(f"Epoch: {epoch} Loss: {loss}")
```
